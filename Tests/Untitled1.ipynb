```python
import pandas as pd
faits_divers = pd.read_csv('Data/data_faits_divers.csv')
#os.chdir("..")
```

```python
import spacy
from tqdm import tqdm


nlp = spacy.load("fr_core_news_md", disable=["ner", "parser", "tagger"])

texts = faits_divers["texte_clean"].tolist()

token_lists = []
for doc in tqdm(nlp.pipe(texts, batch_size=30), desc="Tokenisation",total=len(texts)):
    tokens = [token.text.lower() for token in doc if not token.is_punct and not token.is_space]
    token_lists.append(tokens)

faits_divers["tokens"] = token_lists
```

```python
avant_2017 = faits_divers[faits_divers['annee']<=2017]
apres_2017 = faits_divers[faits_divers['annee']>2017]
```

```python
pre_token = avant_2017['tokens']
post_token = apres_2017['tokens']
```

```python
from gensim.models import Word2Vec

pre_model = Word2Vec(sentences=pre_token, vector_size=300, window=10, min_count=500, workers=5)
```

```python
post_model = Word2Vec(sentences=post_token, vector_size=300, window=10, min_count=500, workers=5)
```

```python
from scipy import spatial
from scipy.spatial import KDTree
import numpy as np


pre_words = list(pre_model.wv.index_to_key)

pre_vectors = np.array([pre_model.wv[word] for word in pre_words])

pre_tree = KDTree(pre_vectors)
```

```python
post_words = list(post_model.wv.index_to_key)

post_vectors = np.array([post_model.wv[word] for word in post_words])

post_tree = KDTree(post_vectors)
```

```python

query_words = ['victime','coupable','suspect']
words_plus_neighbors = []
for word in query_words:
    if word in pre_words:
        i=pre_words.index(word)
        distances, indices = pre_tree.query(pre_vectors[i], k=11)
        nearest_neighbors = [pre_words[j] for j in indices if j!= i]
        print(f"Nearest neighbors to '{word}': {nearest_neighbors}")
        for j in indices:
            if j!= i:
                words_plus_neighbors.append(pre_words[j])
        
    else:
        print(f"'{word}' not found in the vocabulary!")
    
```

```python
query_words = ['victime','coupable','suspect']
words_plus_neighbors = []
for word in query_words:
    if word in post_words:
        i=post_words.index(word)
        distances, indices = post_tree.query(post_vectors[i], k=11)
        nearest_neighbors = [post_words[j] for j in indices if j!= i]
        print(f"Nearest neighbors to '{word}': {nearest_neighbors}")
        for j in indices:
            if j!= i:
                words_plus_neighbors.append(post_words[j])
        
    else:
        print(f"'{word}' not found in the vocabulary!")
    
```
